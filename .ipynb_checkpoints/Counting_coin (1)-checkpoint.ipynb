```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread("/content/a.jpg")
img[100:500, 250:350]
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
blur = cv.GaussianBlur(gray, (13, 13), 0)
plt.imshow(blur   , cmap='gray')
```

```python
canny = cv.Canny(blur, 30, 100, 10)
plt.imshow(canny, cmap='gray')
```

```python
dilated = cv.dilate(canny, (1, 1), iterations = 3)
plt.imshow(dilated, cmap='gray')
```

```python
(cnt, heirarchy) = cv.findContours(dilated.copy(), cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE)
rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
cv.drawContours(rgb, cnt, -1, (0,0,255), 2)

plt.imshow(rgb)
print ('number ::: ', len(cnt))
```
