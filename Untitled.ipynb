## IMPORT Libraries

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

# PreProcessing

```python
# read image
img = plt.imread("1 (7).jpg")
plt.imshow(img)
```

## GRAY

```python

img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap="gray")
```

## Blur

```python
img_blur = cv.GaussianBlur(img_gray, (7, 7), 2)
plt.imshow(img_blur, cmap="gray")
```

## Canny

```python
img_canny = cv.Canny(img_gray, 15, 250, 2 )
plt.imshow(img_canny, cmap="gray")
```

## dilated

```python
img_dilated = cv.dilate(img_canny, (1, 1), iterations = 2)
plt.imshow(img_dilated, cmap="gray")
```

## Final :: findContours

```python
# get copy from img_dilated
# cv.RETR_EXTERNAL :: method for outermost corner
"""
cv.CHAIN_APPROX_NONE ::

stores absolutely all the contour points. That is, any 2 subsequent points (x1,y1) and (x2,y2)
of the contour will be either horizontal, vertical or diagonal neighbors, that is, max(abs(x1-x2),abs(y2-y1))==1.

"""

(cnt, heirarchy) = cv.findContours(img_dilated.copy(), cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE)
rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
# draw contours on image for u see
cv.drawContours(rgb, cnt, -1, (255,0,0), 3)



plt.imshow(rgb)
print ('number ::: ', len(cnt))
```
